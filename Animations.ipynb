#### Creating small Animation

```python
os.mkdir(r"C:\Users\91912\OneDrive\Documents\Animation")
```

```python
x1 = -1
y1 = -1
def mouse_tracking(event,x,y,flags,param):
    global x1,y1
    if event == 1:
        x1 = x
        y1 = y   
    elif event == 2:
        cv2.rectangle(black_image,(x1,y1),(x,y),(100,150,200),2)
```

```python
cv2.namedWindow("mouse Tracking")

cv2.setMouseCallback("mouse Tracking",mouse_tracking)

black_image = np.zeros((500,500,3),dtype=np.uint8)

c = 0
while True:

    cv2.imshow("mouse Tracking",black_image)

    if cv2.waitKey(1) & 255 == ord("s"):
        cv2.imwrite(r"C:\Users\91912\OneDrive\Documents\Animation\{}.jpg".format(c),black_image)
        print("Saved Successfully")
        c+=1

    if cv2.waitKey(1) & 255 == ord("q"):
        break

cv2.destroyAllWindows()
```
