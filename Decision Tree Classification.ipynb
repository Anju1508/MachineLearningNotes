```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn 
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
fv,cv = make_classification(n_samples = 100,
                            n_features = 2,
                            n_informative = 2,
                            n_redundant = 0,
                            n_repeated = 0,
                            n_classes = 2,
                            n_clusters_per_class = 1,
                            weights = [0.5,0.5],
                            class_sep = 0.8,
                            random_state = 3)
```

```python
sns.scatterplot(x = fv[:,0],y =fv[:,1],hue = cv)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth =None,min_samples_leaf = 10)
```

```python
dtc.fit(fv,cv)
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(dtc,filled = True)
plt.show()
```

```python
import mlxtend
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(clf = dtc,X=fv,y = cv,legend = 2)
plt.show()
```

```python
data = sns.load_dataset("iris")
data.head()
```

```python
fv = data.iloc[:,:-1]
cv = data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(fv,cv,test_size = 0.2,stratify = cv)
```

```python
dt = DecisionTreeClassifier(max_depth =1)
dt.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (10,10))
tree.plot_tree(dt,filled = True,class_names=cv.unique(),feature_names = fv.columns)
plt.show()
```
