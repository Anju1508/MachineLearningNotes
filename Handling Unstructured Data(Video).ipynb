```python
import cv2
import numpy as np
import os
import pandas as pd
```

### How to play video in OpenCV (cv2.VideoCapture)

```python
vid = cv2.VideoCapture(r"C:\Users\91912\Downloads\WhatsApp Video 2024-11-26 at 12.11.28 PM.mp4")

while True:
    succ,img = vid.read()  # it gives two outputs(1-frames of video and 2-boolean value according into presence of frame in vid)

    if succ == False:
        break

    cv2.imshow("video",img)
    
    if cv2.waitKey(100) & 255 == ord("a"):
        break

cv2.destroyAllWindows()
```

#### Converting Video to Grey Scale

```python
vid = cv2.VideoCapture(r"C:\Users\91912\Downloads\WhatsApp Video 2024-11-26 at 12.11.28 PM.mp4")

while True:
    succ,img = vid.read()

    if succ == False:
        break
        
    img1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Convert bgr image to grey scale
    cv2.imshow("video",img)
    cv2.imshow("video_grey",img1)
    
    if cv2.waitKey(100) & 255 == ord("a"):
        break

cv2.destroyAllWindows()
```

#### Converting Video to B,G,R shades

```python
vid = cv2.VideoCapture(r"C:\Users\91912\Downloads\WhatsApp Video 2024-11-26 at 12.11.28 PM.mp4")

while True:
    succ,img = vid.read()

    if succ == False:
        break

    b,g,r = cv2.split(img)
    z = np.zeros(b.shape,dtype = np.uint8)

    img1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    cv2.imshow("Grey",img1)
    cv2.imshow("blue",cv2.merge([b,z,z]))
    cv2.imshow("Green",cv2.merge([z,g,z]))
    cv2.imshow("Red",cv2.merge([z,z,r]))
    cv2.imshow("video",img)
    
    if cv2.waitKey(100) & 255 == ord("a"):
        break

cv2.destroyAllWindows()
```

#### Converting Videos to Tabular Structure

```python
os.mkdir(r"C:\Users\91912\OneDrive\Documents\Animals\Cat")
os.mkdir(r"C:\Users\91912\OneDrive\Documents\Animals\Dog")
os.mkdir(r"C:\Users\91912\OneDrive\Documents\Animals\Elephant")
```

```python
l = os.listdir(r"C:\Users\91912\OneDrive\Documents\Animals")
l
```

```python
l1 = []
for i in l:
    if i.endswith(".mp4"):
        l1.append(i)
```

```python
l1
```

```python
for video in l1:
    vid = cv2.VideoCapture(r"C:\Users\91912\OneDrive\Documents\Animals\{}".format(video))
    c = 0
    while True:
        succ,img = vid.read()

        if succ == False:
            break
        cv2.imwrite(r"C:\Users\91912\OneDrive\Documents\Animals\{}\{}{}.jpg".format(video.split(".")[0],video.split(".")[0],c),img)
        c += 1
        
        cv2.imshow("vid",img)
        if cv2.waitKey(100) & 255 == ord("a"):
            break

    cv2.destroyAllWindows()
```

```python
l2 = []
for i in l:
    if not i.endswith(".mp4"):
        l2.append(i)
```

```python
l2
```

```python
features = []
class_labels = []
for i in l2:
    for images in os.listdir(r"C:\Users\91912\OneDrive\Documents\Animals\{}".format(i)):
        img = cv2.imread(r"C:\Users\91912\OneDrive\Documents\Animals\{}\{}".format(i,images),0)
        img = cv2.resize(img,(50,50)) # resize the image
        img = img.flatten()  # Falttening array
        features.append(img)
        class_labels.append(i)
```

```python
df = pd.DataFrame(features)
```

```python
df["Class_labels"] = class_labels
```
