```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn 
from sklearn.datasets import make_classification,make_regression
from sklearn.model_selection import train_test_split,cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
fv,cv = make_classification(n_samples = 800,
                            n_features = 2,
                            n_informative = 2,
                            n_redundant = 0,
                            n_repeated = 0,
                            n_classes = 2,
                            n_clusters_per_class = 1,
                            weights = [0.5,0.5],
                            class_sep = 1.8,
                            random_state = 5)
```

```python
sns.scatterplot(x = fv[:,0],y =fv[:,1],hue = cv)
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(fv,cv,test_size =0.2,stratify = cv)
```

```python
rf = RandomForestClassifier(n_estimators = 4,max_features = "sqrt",bootstrap = True,max_samples = 100)
```

```python
rf
```

```python
rf.fit(x_train,y_train)
```
